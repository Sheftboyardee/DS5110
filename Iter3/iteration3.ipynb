```python
#making standardized naming convention for columns
sales_data.columns = (sales_data.columns.str.strip().str.lower().str.replace(' ', '_'))
sales_data.head()
```

```python
#reformatting the date column
sales_data['date'] = pd.to_datetime(sales_data['date'])
sales_data['date'] = sales_data['date'].dt.strftime('%m/%d/%Y')
sales_data.head()
```

```python
#normalizing numeric cols
numeric_cols= ['inventory_level', 'units_sold', 'units_ordered', 'price',
                'discount', 'competitor_pricing', 'demand']
scaler = MinMaxScaler()
sales_data[numeric_cols] = scaler.fit_transform(sales_data[numeric_cols])
sales_data.head()
```

```python
#sorting
sales_data = sales_data.sort_values(['store_id', 'product_id', 'date'])
sales_data
```

```python
#extracting date info
sales_data['date'] = pd.to_datetime(sales_data['date'])
sales_data['year'] = sales_data['date'].dt.year
sales_data['month'] = sales_data['date'].dt.month
sales_data['day'] = sales_data['date'].dt.day
sales_data['day_of_week'] = sales_data['date'].dt.dayofweek
sales_data['week_of_year'] = sales_data['date'].dt.isocalendar().week.astype(int)
sales_data['is_weekend'] = sales_data['day_of_week'].isin([5, 6]).astype(int)
sales_data.head()
```

```python
#feature engineering for cyclical (time) data using sin and cos
sales_data['month_sin'] = np.sin(2 * np.pi * sales_data['month'] / 12)
sales_data['month_cos'] = np.cos(2 * np.pi * sales_data['month'] / 12)
sales_data['dayofweek_sin'] = np.sin(2 * np.pi * sales_data['day_of_week'] / 7)
sales_data['dayofweek_cos'] = np.cos(2 * np.pi * sales_data['day_of_week'] / 7)
sales_data.head()
```

```python
#features for pricing and sales promotions
sales_data['price_discount_ratio'] = (sales_data['discount'] / sales_data['price']).fillna(0)
sales_data['price_vs_competitor'] = ((sales_data['price'] - sales_data['competitor_pricing']) / sales_data['competitor_pricing']
).fillna(0)
sales_data['was_discounted'] = (sales_data['discount'] > 0).astype(int)
sales_data.head()
```

```python
sales_data.to_csv("sales_data_modified.csv", index = False)
```
